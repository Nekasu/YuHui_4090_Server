```python
# 将一张图像转换为频域, 并打印
from fileinput import filename
import PIL.Image as Image
import numpy as np
import matplotlib.pyplot as plt
filename = 'bamboo_large_leaves.png'
pth = '/home/zxt/Python_area/for_wu_ding_minecraft_modules/datas/1_origin_data/' + filename
# 读取图像
img = Image.open(pth)
img = img.convert('L')
img = np.array(img)
# 转换为频域
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))
# 打印
plt.imshow(magnitude_spectrum, cmap='gray')
plt.show()
```

```python
# 下采样频域图像
rows, cols = img.shape
crow, ccol = rows//2, cols//2
fshift[crow-30:crow+30, ccol-30:ccol+30] = 0

# 转换为空域
f_ishift = np.fft.ifftshift(fshift)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)
# 打印
plt.imshow(img_back, cmap='gray')
```

```python
# 函数定义, 将32*32的风格图像放大到512*512, 原图中的每个像素向右下扩散16*16个像素
def super_resolution_to_512(img):
    # print(img_torch.size())
    img_512 = np.zeros((512, 512))
    # print(img_512.size())
    for j in range(0, 32):
        for k in range(0, 32):
            img_512[16*j:16*(j+1), 16*k:16*(k+1)] = img[j, k]
    # print(img_512.size())
    return img_512
```

```python
#上采样频域图像
fshift = super_resolution_to_512(magnitude_spectrum)
# 转换为空域
f_ishift = np.fft.ifftshift(fshift)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)
# 打印
plt.imshow(img_back, cmap='gray')
```
